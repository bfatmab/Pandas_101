```python
# "who" değişkenini daatframe'den düşürün 
# "axis=1" -> kolonlardan düşür yani sil demektir
df.drop("who",axis=1, inplace =True)
df.head(10)
#diğer yazımlaro: 
df= df.drop("who", axis=1)
df=df.groupby(columns["who"])
```

```python
# "deck" değişkenindeki boş değerleri "deck" değişkeninin 
#en çok tekerar eden değeri ile doldurunuz.

#öncelikle en çok tekrar eden 
df["deck"]. mode()
type(df["deck"].mode())

#şimdi bunun bir tipine bakalım panda.series çıakcak ve işimize yaramaz
#o sebepten ötürü biz en çok tekrar eden değerin index'ine odaklanalım yani [0]'a 

df["deck"].mode()[0]
df["deck"].fillna(df["deck"].mode()[0], inplace=True)
df["deck"].isnull().sum()

# eğer en sonunda inplace =True denilmezse kalıcı bir sonuç vermez.
```

```python
 # "age" değişkenindeki boş değerleri "age"  değikeninin medyanı ile doldurunuz 
df["age"].median()
df["age"].fillna(df["age"].median(), inplace=True) 
df["age"].isnull().sum()
```

```python
# "survived" değişkeninin Pclass ve Cinsiyet değikenleri kırılımında; 
#sum, count ,mean değerlerini bulunuz 
df.groupby(["pclass","sex"]).agg({"survived": ["sum","count","mean"]})
```

```python
#yolcuları sınıf (Pclass) ve cinsiyet (Sex) temelinde gruplandırarak hayatta kalma (Survived) sayısını, yaş ortalamasını (Age), ve ücret ortalamasını (Fare) hesaplayabiliriz.

df.groupby(['pclass', 'sex']).agg({
    'survived': ['sum', 'count', 'mean'],  # Hayatta kalan yolcuların sayısı, toplam yolcu sayısı, ve hayatta kalma oranı
    'age': ['mean', 'median'],             # Yaş ortalaması ve medyanı
    'fare': ['mean', 'sum']                # Ücret ortalaması ve toplamı
})
```

```python
# 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 verecek bir fonk.yazınız 
#Yazdığımız fonksiyonu kullanarak titanic veri setine age_flag adında bir değişken oluşturunuz.(apply ve lambda ypıları kullanarak) 

#Yol 1:
df["age_flag"]=df["age"].apply(lambda x: int(x<30))
df["age_flag"]

#Yol 2:
#lambda x: <true_case> if <condition> else <false_case>
#df["age_flag"]=df["age"].apply(lambda x: 1 if<30 else 0)
```

```python
#Yukarıdaki örneği comprehension ile yapmak istersek: 
df["age_flag"]=[1 if item<30 else 0 for item in df["age"]]
```

```python
#Seaborn kütüphanesinden içerisinden Tips vei setini tanımlayalım. 
df1 =sns.load_dataset("tips")
df1.head()
```

```python
#Time değişkeninin kategorilerien (Dinner, Lunch) göre "total_bill" değerleinin; 
#toplamını ,min,max ve ort. bulunuz 

#groupby ile :
df1.groupby("time").agg({"total_bill":["sum","min","max","mean"]})
```

```python
#Pivot table ile yapalım aynı soruyu: 
df1.pivot_table("total_bill","time",aggfunc=[np.sum,np.max,np.mean,np.min])
#ya da (or)
#df1.pivot_table("total_bill","time",aggfunc=["sum","max","mean","min"])

#Pivot ile groupby farkı sütün yapısının farklı olması
```

```python
#"day" ve "time" değişkenlerine göre "total_bill" değerlerinin; 
#toplam ,min,max ve ortalamalarını bulunuz 

#groupby ile (with a groupby): 
df1.groupby(["day","time"]).agg({"total_bill":["sum","max","min","mean"]})
```

```python
# Yukarıdaki soruyu pivot table ile yapalım : 
df1.pivot_table("total_bill",["day","time"],aggfunc=["sum","min","max","mean"])
```

```python
#(***)"Lunch"zamanına ve "kadın" müşterilere ait "total_bill"  ve "tip" değerlerinin 
# "day"'e göre toplam,min,max ve ort bulunuz 

df1[(df1["time"]=="Lunch") & (df1["sex"]=="Female")].groupby("day").agg({"total_bill":["sum","min","max","mean"], 
                                                                       "tip":["sum","min","max","mean"]})
```

```python
#Yukarıda "Sat" ve "Sun" satırları boş olduğu görülmektedir."Nan" görmek istemzsek 
df1[(df1["time"]=="Lunch") & (df1["sex"]=="Female")].groupby("day").agg({"total_bill":["sum","min","max","mean"], 
                                                                       "tip":["sum","min","max","mean"]}).query('day not in ["Sun", "Sat"]')

```
