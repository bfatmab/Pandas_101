```python
df.loc[0:3,df.columns.str.contains("yas")].apply(lambda x: x/10).head()
```

```python

df.loc[:, df.columns.str.contains("isim")].apply(lambda x: x.str.upper()).head()
```


Grouping (Gruplama)
(EN)
Grouping is the process of splitting data into groups based on certain criteria. In Pandas, the groupby() function is used to group data based on one or more columns. This allows for the analysis of data within each group separately, making it easier to calculate statistics for each category or subset.

(TR)
Gruplama, verileri belirli kriterlere göre gruplara ayırma işlemidir. Pandas'ta groupby() fonksiyonu kullanılarak veriler bir veya daha fazla sütuna göre gruplandırılır. Bu, her grup içinde veriyi ayrı ayrı analiz etmeyi sağlar ve her kategori veya alt küme için istatistik hesaplamayı kolaylaştırır.
